```python
if(dataset["Mileage"].dtype=='O'):
    print("qual")
else:
    print("quan")
```

```python
quan=[]
qual=[]
for columnName in dataset.columns:
    #print(columnName)
    if(dataset[columnName].dtype=='O'):
       #print("qual")
       qual.append(columnName)
    else:
        #print("quan")
        quan.append(columnName)
```

```python
def quanQual(dataset):
        quan=[]
        qual=[]
        for columnName in dataset.columns:
            #print(columnName)
            if(dataset[columnName].dtype=='O'):
               #print("qual")
               qual.append(columnName)
            else:
                #print("quan")
                quan.append(columnName)
        return quan,qual    
```

```python
quanQual(dataset)
```

```python
descriptive=pd.DataFrame(index=["mean","median","mode"],columns=quan)
descriptive
```

```python
descriptive=pd.DataFrame(index=["mean","median","mode"],columns=quan)
for columnName in quan:
    descriptive[columnName]["mean"]=dataset[columnName].mean()
    descriptive[columnName]["median"]=dataset[columnName].median()
    descriptive[columnName]["mode"]=dataset[columnName].mode()[0]
descriptive
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#how many vechicles need maintenance or not
dataset['Need_Maintenance'].value_counts().plot(kind='bar')
plt.title("Need Maintenance Count")
plt.xlabel("Need Maintenance")
plt.ylabel("Count")
plt.show()
```

```python
#Mileage Distribution
plt.hist(dataset['Mileage'], bins=20)
plt.title("Mileage Distribution")
plt.xlabel("Mileage")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x=dataset['Vehicle_Age'])
plt.title("Vehicle Age Distribution")
plt.show()
```

# Bivariate:

```python
#mileage vs Need_maintenace
sns.boxplot(x='Need_Maintenance', y='Mileage', data=dataset)
plt.title("Mileage vs Need Maintenance")
plt.show()
```

```python
#Reported _issue vs Need_maintenance
sns.barplot(x='Need_Maintenance', y='Vehicle_Age', data=dataset)
plt.title("Vehicle Age vs Need Maintenance")
plt.show()
```
